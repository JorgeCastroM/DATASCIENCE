## Crear las tablas

### Subtask:
Generar y ejecutar sentencias SQL para crear las tablas en la base de datos.


**Reasoning**:
Generate and execute SQL statements to create tables based on the defined schema.



```python
cursor = conn.cursor()

for table_name, table_info in schema.items():
    columns = []
    for col in table_info["columns"]:
        col_definition = f'"{col["name"]}" {col["type"]}'
        if col.get("primary_key"):
            col_definition += " PRIMARY KEY"
        columns.append(col_definition)

    create_table_sql = f'CREATE TABLE IF NOT EXISTS "{table_name}" ({", ".join(columns)});'
    print(f"Executing SQL: {create_table_sql}")
    cursor.execute(create_table_sql)

conn.commit()
print("Tables created successfully.")
```

This error often occurs in SQLite when trying to add foreign keys to existing tables using `ALTER TABLE ADD FOREIGN KEY`. A common workaround is to recreate the table with the foreign key constraint included in the `CREATE TABLE` statement, copy the data, drop the old table, and rename the new table.



```python
cursor = conn.cursor()

# Temporarily disable foreign key checks
cursor.execute("PRAGMA foreign_keys=OFF;")

# Define the relationships with foreign key constraints
relationships = {
    "Distritos": [
        {"column": "CANTON", "references_table": "Cantones", "references_column": "CÓDIGOS DE CANTONES"}
    ],
    "Regiones": [
        {"column": "COD_DISTRITO", "references_table": "Distritos", "references_column": "CÓDIGO DISTRITO"},
        {"column": "COD_CANTÓN", "references_table": "Cantones", "references_column": "CÓDIGOS DE CANTONES"}
    ],
    "Provincia": [
        {"column": "COD_DISTRITO", "references_table": "Distritos", "references_column": "CÓDIGO DISTRITO"},
        {"column": "COD_CANTÓN", "references_table": "Cantones", "references_column": "CÓDIGOS DE CANTONES"}
    ]
}

try:
    for table_name, rel_list in relationships.items():
        # Get the schema for the current table
        table_info = schema[table_name]
        columns_definitions = []
        for col in table_info["columns"]:
            col_definition = f'"{col["name"]}" {col["type"]}'
            if col.get("primary_key"):
                col_definition += " PRIMARY KEY"
            columns_definitions.append(col_definition)

        # Add foreign key definitions to the CREATE TABLE statement
        for rel in rel_list:
            fk_definition = f'FOREIGN KEY ("{rel["column"]}") REFERENCES "{rel["references_table"]}"("{rel["references_column"]}")'
            columns_definitions.append(fk_definition)

        # Create a new table with the foreign key constraints
        new_table_name = f"{table_name}_new"
        create_new_table_sql = f'CREATE TABLE "{new_table_name}" ({", ".join(columns_definitions)});'
        print(f"Executing SQL: {create_new_table_sql}")
        cursor.execute(create_new_table_sql)

        # Copy data from the old table to the new table
        columns_list = [f'"{col["name"]}"' for col in table_info["columns"]]
        copy_data_sql = f'INSERT INTO "{new_table_name}" ({", ".join(columns_list)}) SELECT {", ".join(columns_list)} FROM "{table_name}";'
        print(f"Executing SQL: {copy_data_sql}")
        cursor.execute(copy_data_sql)

        # Drop the old table
        drop_old_table_sql = f'DROP TABLE "{table_name}";'
        print(f"Executing SQL: {drop_old_table_sql}")
        cursor.execute(drop_old_table_sql)

        # Rename the new table to the original table name
        rename_table_sql = f'ALTER TABLE "{new_table_name}" RENAME TO "{table_name}";'
        print(f"Executing SQL: {rename_table_sql}")
        cursor.execute(rename_table_sql)

    # Re-enable foreign key checks
    cursor.execute("PRAGMA foreign_keys=ON;")

    conn.commit()
    print("Foreign key constraints added successfully by recreating tables.")

except sqlite3.Error as e:
    print(f"An error occurred: {e}")
    conn.rollback() # Rollback changes if an error occurs
    # Re-enable foreign key checks in case of error
    cursor.execute("PRAGMA foreign_keys=ON;")
```
