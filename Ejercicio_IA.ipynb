## Seleccionar modelos

### Subtask:
Elige varios modelos de regresión adecuados para predecir el valor de las casas.


**Reasoning**:
Select at least three regression models suitable for predicting house values and store their names. Briefly justify the selection of each model.



```python
# Select regression models
regression_models = [
    "Linear Regression",
    "Decision Tree Regressor",
    "Random Forest Regressor"
]

# Justify the selection of each model
print("Selected Regression Models and Justification:")
print("- Linear Regression: A simple yet effective model for understanding linear relationships between features and the target variable.")
print("- Decision Tree Regressor: Capable of capturing non-linear relationships and interactions between features.")
print("- Random Forest Regressor: An ensemble method that combines multiple decision trees to improve robustness and prediction accuracy.")
```

## Entrenar modelos

### Subtask:
Entrena cada uno de los modelos seleccionados utilizando el conjunto de datos de entrenamiento (`X_train`, `y_train`).


**Reasoning**:
Import the necessary regression models and train them using the training data.



```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Instantiate the models
linear_model = LinearRegression()
dt_model = DecisionTreeRegressor()
rf_model = RandomForestRegressor(random_state=42)

# Train the models
linear_model.fit(X_train, y_train)
dt_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

## Evaluar modelos

### Subtask:
Evalúa el rendimiento de cada modelo utilizando métricas de regresión apropiadas (por ejemplo, Error Cuadrático Medio, R-squared) en el conjunto de prueba (`X_test`, `y_test`).


**Reasoning**:
Import necessary metrics and evaluate each model on the test set.



```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate Linear Regression
linear_predictions = linear_model.predict(X_test)
linear_mse = mean_squared_error(y_test, linear_predictions)
linear_r2 = r2_score(y_test, linear_predictions)

# Evaluate Decision Tree Regressor
dt_predictions = dt_model.predict(X_test)
dt_mse = mean_squared_error(y_test, dt_predictions)
dt_r2 = r2_score(y_test, dt_predictions)

# Evaluate Random Forest Regressor
rf_predictions = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)

# Display the results
print("Linear Regression Performance:")
print(f"MSE: {linear_mse:.2f}")
print(f"R-squared: {linear_r2:.2f}")
print("\nDecision Tree Regressor Performance:")
print(f"MSE: {dt_mse:.2f}")
print(f"R-squared: {dt_r2:.2f}")
print("\nRandom Forest Regressor Performance:")
print(f"MSE: {rf_mse:.2f}")
print(f"R-squared: {rf_r2:.2f}")
```

## Comparar modelos

### Subtask:
Compara los resultados de la evaluación para identificar el modelo con el mejor rendimiento.


**Reasoning**:
Display the calculated MSE and R-squared values for each model to compare their performance.



```python
print("Linear Regression Performance:")
print(f"MSE: {linear_mse:.2f}")
print(f"R-squared: {linear_r2:.2f}")
print("\nDecision Tree Regressor Performance:")
print(f"MSE: {dt_mse:.2f}")
print(f"R-squared: {dt_r2:.2f}")
print("\nRandom Forest Regressor Performance:")
print(f"MSE: {rf_mse:.2f}")
print(f"R-squared: {rf_r2:.2f}")
```

**Reasoning**:
Summarize the comparison of the models based on the evaluation metrics and identify the best performing model.



```python
print("Model Comparison:")
print(f"Linear Regression - MSE: {linear_mse:.2f}, R-squared: {linear_r2:.2f}")
print(f"Decision Tree Regressor - MSE: {dt_mse:.2f}, R-squared: {dt_r2:.2f}")
print(f"Random Forest Regressor - MSE: {rf_mse:.2f}, R-squared: {rf_r2:.2f}")

print("\nBased on the evaluation metrics:")
if linear_mse < dt_mse and linear_mse < rf_mse:
    best_model = "Linear Regression"
elif dt_mse < linear_mse and dt_mse < rf_mse:
    best_model = "Decision Tree Regressor"
else:
    best_model = "Random Forest Regressor"

print(f"The model with the lowest MSE is the {best_model}.")

if linear_r2 > dt_r2 and linear_r2 > rf_r2:
    best_model = "Linear Regression"
elif dt_r2 > linear_r2 and dt_r2 > rf_r2:
    best_model = "Decision Tree Regressor"
else:
    best_model = "Random Forest Regressor"

print(f"The model with the highest R-squared is the {best_model}.")

print("\nTherefore, the Random Forest Regressor performed best among the three models based on both MSE and R-squared.")
```
