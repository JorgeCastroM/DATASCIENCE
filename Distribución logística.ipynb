## Función logística

####Parametros: **loc** (determinación en 0 para gráfica), **scale**(desviación estandar), y **size**(la forma de la matriz de retorno)

```python
import numpy as np
```

```python
x = np.random.logistic(loc= 1, scale=2, size= (2,2))
```

```python
print(x)
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import logistic

# Generate points for the x-axis
x_values = np.linspace(x.min(), x.max(), 100)

# Calculate the PDF for each point
pdf_values = logistic.pdf(x_values, loc=1, scale=2) # Using the same loc and scale as before

# Plot the PDF
plt.plot(x_values, pdf_values, color='red', label='Logistic PDF')
plt.title('Continuous Logistic Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

*## Se usa para modelar eventos y predecir mejor eventos extremos, aprovechando sus colas más pesadas en comparación con la distribución normal.*
